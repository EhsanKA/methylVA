## Generate random data and save it (negative sample)

```python
import pandas as pd
import pickle
import argparse
from methylVA.utils.common import load_config
from sklearn.model_selection import train_test_split
import numpy as np


```

```python
n_features = 1000
n_samples_train = 900
n_samples_test = 100

train_data_path = f"../data/random_data/train_data_{n_samples_train}_{n_features}.pkl"
test_data_path = f"../data/random_data/test_data_{n_samples_test}_{n_features}.pkl"

train_metadata_path = f"../data/random_data/train_metadata_{n_samples_train}_{n_features}.pkl"
test_metadata_path = f"../data/random_data/test_metadata_{n_samples_test}_{n_features}.pkl"
```

```python
def generate_random_data(n_rows=100, n_cols=1000, seed=42):
    np.random.seed(seed)
    data = np.random.rand(n_rows, n_cols)
    data_df = pd.DataFrame(data, columns=[f"col_{i+1}" for i in range(n_cols)])
    
    labels = np.random.randint(0, 10, size=n_rows)
    metadata_df = pd.DataFrame({'labels_encoded': labels})
    
    return data_df, metadata_df
```

```python
train_data, train_metadata = generate_random_data(n_rows=n_samples_train,
                                                  n_cols=n_features,
                                                  seed=42
                                                 )

test_data, test_metadata = generate_random_data(n_rows=n_samples_test,
                                                n_cols=n_features,
                                                seed=43
                                               )

```

```python
train_data
```

```python
train_data = train_data.reset_index()
test_data = test_data.reset_index()

train_data.to_pickle(train_data_path)
test_data.to_pickle(test_data_path)

train_metadata.to_pickle(train_metadata_path)
test_metadata.to_pickle(test_metadata_path)
```
